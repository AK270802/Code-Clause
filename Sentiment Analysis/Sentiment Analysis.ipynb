# Sentiment Analysis

### Introduction to Sentiment Analysis using Logistic Regression
Sentiment analysis, also known as opinion mining, is a natural language processing technique that aims to determine the sentiment or emotional polarity expressed in a piece of text. Logistic regression is a widely used machine learning algorithm that can be applied to sentiment analysis tasks. It is particularly effective in binary classification problems, such as determining whether a given text expresses a positive or negative sentiment.
By employing logistic regression for sentiment analysis, we can automatically analyze and classify the sentiment expressed in text data. This can be valuable for various applications, including social media monitoring, customer feedback analysis, and market research.

```python
stuff_to_be_removed = list(stopwords.words('english'))+list(punctuation)
stemmer = LancasterStemmer()
corpus = df['text'].tolist()
print(len(corpus))
print(corpus[0])
```

```python
final_corpus = []
final_corpus_joined = []
for i in df.index:
    text = re.sub('[^a-zA-Z]', ' ', df['text'][i])
    text = text.lower()
    text=re.sub("&lt;/?.*?&gt;"," &lt;&gt; ",text)
    text=re.sub("(\\d|\\W)+"," ",text)
    text = text.split()
    lem = WordNetLemmatizer()
    text = [lem.lemmatize(word) for word in text  if not word in stuff_to_be_removed] 
    text1 = " ".join(text)
    final_corpus.append(text)
    final_corpus_joined.append(text1)
```

```python
data_cleaned = pd.DataFrame()
data_cleaned["text"] = final_corpus_joined
data_cleaned["sentiment"] = df["sentiment"].values
```

```python
data_eda = pd.DataFrame()
data_eda['text'] = final_corpus
data_eda['sentiment'] = df['sentiment'].values
data_eda.head()
```

```python
positive = data_eda[data_eda['sentiment'] == 4]
positive_list = positive['text'].tolist()
negative = data_eda[data_eda['sentiment'] == 0]
negative_list = negative['text'].tolist()
```

```python
positive_all = " ".join([word for sent in positive_list for word in sent ])
negative_all = " ".join([word for sent in negative_list for word in sent ])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
xt = tfidf.fit_transform(data_cleaned['text'])
y = data_cleaned['sentiment']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(xt, 
                                                    y, 
                                                    test_size=0.33, 
                                                    random_state=42,
                                                    stratify = y)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)
accuracy_score(y_train,y_train_pred)*100
```

### Conclusion:
In conclusion, logistic regression is a valuable tool in sentiment analysis, enabling us to analyze and classify the sentiment expressed in text data. By leveraging its binary classification capabilities, we can determine whether a given text exhibits a positive or negative sentiment. Logistic regression, when combined with appropriate preprocessing techniques and feature extraction methods, provides a robust framework for sentiment analysis tasks.

However, it's important to note that sentiment analysis is a challenging task due to the nuances and subjectivity of language. While logistic regression can yield good results, it may not capture the full complexity of sentiment expression in some cases. It's essential to continuously improve and fine-tune the model by exploring alternative algorithms, incorporating domain-specific knowledge, and considering more advanced techniques like neural networks or ensemble methods.
